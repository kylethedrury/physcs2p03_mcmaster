```python
# Must find ratio between magnetic field B and current for each current spot (target 2, foil1, coll12, coll23)

cyc_freq = 27200000
mp = 1.67e-27
e = 1.6e-19 

def field_to_current_ratio(f,m,e,I):
    return 2*f*math.pi*m/(e*I)
```

```python
foil1_Bratio = []

for i in range(len(foil1)):
    foil1_Bratio.append(field_to_current_ratio(cyc_freq, mp, e, foil1[i]))
    
print(foil1)
print(foil1_Bratio) 

plt.plot(supply_current, foil1_Bratio, 'x')
plt.ylim(0,4)
plt.show()
```

```python
target2_Bratio = []

for i in range(len(target2)):
    target2_Bratio.append(field_to_current_ratio(cyc_freq, mp, e, target2[i]))
    
print(target2)
print(target2_Bratio) 

plt.plot(supply_current, target2_Bratio, 'x')
plt.ylim(0,4)
plt.show()
```

```python
coll12_Bratio = []

for i in range(len(coll12)):
    coll12_Bratio.append(field_to_current_ratio(cyc_freq, mp, e, coll12[i]))
    
print(coll12)
print(coll12_Bratio) 

plt.plot(supply_current, coll12_Bratio, 'x')
plt.ylim(0,19)
plt.show()
```

```python
coll23_Bratio = []

for i in range(len(coll12)):
    coll23_Bratio.append(field_to_current_ratio(cyc_freq, mp, e, coll23[i]))
    
print(coll23)
print(coll23_Bratio) 

plt.plot(supply_current, coll23_Bratio, 'x')
plt.ylim(0,19)
plt.show()
```
